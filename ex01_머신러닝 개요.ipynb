```python
# Question1
# 문제를 지정하고, 데이터 직접 만들기
# 규칙 : 큰 수 + 두 수의 차
# 22 ? 7 = 37
```

```python
# 데이터 수집
```

```python
list1 = [] #저장 용도로 만들어진 공간
for i in range(100):
    for j in range(100):
        result = (i + (i-j)) if i > j else (j + (j-i))
        list1.append([i, j, result])
        
list1
```

```python
# 데이터 전처리
# 머신러닝 모델에 학습할 형식으로 변환 
# DataFrame, Series, np.array 많이 사용 => 계산 가능
```

```python
import pandas as pd
data1 = pd.DataFrame(list1, columns=['N1','N2','Result'])
data1.head()
```

```python
# 빠른 연산을 위해 벡터 표현방식을 사용
# 문제(X) : 대문자 : 데이터가 2줄 이상 (2차원 이상)
# 정답(y) : 소문자 : 데이터가 1줄 (1차원)

X = data1.loc[:,'N1':'N2']
y = data1.loc[:,'Result']
```

```python
# 학습
# 머신러닝 모델이 규칙을 찾게 해주기

# 사용할 모델 불러오기
from sklearn.ensemble import RandomForestRegressor
random_forest = RandomForestRegressor()

# 규칙 찾게하기
# 가지고 있는 전체 데이터를 살펴보고 문제로 정답을 도출할 수 있는 규칙 1가지 찾기
# 여러가지 규칙 중 제일 적합한 1가지를 가지고 온 상태
random_forest.fit(X, y)
```

```python
# 예측하기
# 찾은 규칙을 토대로 입력한 데이터의 정답 예측해보기
#random_forest.predict([[22,7]])
# 이중 리스트를 사용하는 이유
# X 형식과 같은 형식으로 데이터를 입력
random_forest.predict(pd.DataFrame([[22,7]], columns=['N1','N2']))
```
